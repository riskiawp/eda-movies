```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('downloads/netflix_titles.csv/netflix_titles.csv')
```

```python
dataset.head()
```

```python
dataset.describe(include='all')
```

```python
dataset = dataset.fillna('Unknown')
```

```python
dataset.head()
```

```python
print(dataset['type'].value_counts())
```

```python
#compare movie and tv show
plt.pie(dataset['type'].value_counts(),
       labels = ['Movie', 'Tv Show'],
       shadow = True,
       explode = (0,0.07),
       colors = ['#ffdb58', '#3895D3'],
       autopct = '%1.1f%%')
plt.show()
```

```python
from wordcloud import WordCloud
wc = WordCloud(background_color='white', max_words=500)
fig = plt.figure(figsize = (8,5))
wc.generate(str(dataset['listed_in']))
plt.imshow(wc, interpolation = 'bilinear')
plt.title('Word of genre movie or tv show')
plt.axis('off')
plt.show()
```

```python
from wordcloud import WordCloud
wc = WordCloud(background_color='white', max_words=1000)
fig = plt.figure(figsize = (8,5))
wc.generate(str(dataset['description']))
plt.imshow(wc, interpolation = 'bilinear')
plt.title('Word of describe movie or tv show')
plt.axis('off')
plt.show()
```

```python
print(dataset['rating'].value_counts())
```

```python
dataset['rating'].value_counts().sort_values(ascending = False).plot(kind = 'bar', 
                                                                     figsize = (8,5), 
                                                                     color = '#3895D3')
plt.title('Movie and Tv Show Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
dataset['rating'].value_counts().sort_values(ascending = False).plot(kind = 'barh', 
                                                                     figsize = (8,5), 
                                                                     color = '#3895D3')
plt.title('Movie and Tv Show Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
